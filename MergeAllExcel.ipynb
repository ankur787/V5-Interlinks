```python
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import glob
import os
import re

pd.set_option('display.max_columns', None)
```

```python
server = 'LAPTOP-5PD425U8'
database = 'V5Interlinks_Australia'
driver = 'SQL Server Native Client 11.0'
connection = f'mssql://@{server}/{database}?driver={driver}'

engine = create_engine(connection,pool_size=10, max_overflow=20)
conn = engine.connect()
```

```python
files_xlsx = glob.glob("*.xlsx")
print('There are', len(files_xlsx), 'Data:')
for x in files_xlsx:
    print(x+ ' ', end= ',')
```

```python
df = pd.DataFrame()
```

```python
for f in files_xlsx:
    print( f)
    data = pd.read_excel(f)
    data.columns = data.columns.str.strip()
    df   = df.append(data)
```

```python
s = []
for i in df['StreetAddress']:
    try:
        a= re.sub("^\s+|\s+$", "", i, flags=re.UNICODE)
        a= " ".join(re.split("\s+", a, flags=re.UNICODE))
        s.append(a)
    except:
        s.append(np.nan)
    
df['StreetAddress'] = s

t = []
for i in df['Title']:
    try:
        b= re.sub("^\s+|\s+$", "", i, flags=re.UNICODE)
        b= " ".join(re.split("\s+", b, flags=re.UNICODE))
        t.append(b)
    except:
        t.append(np.nan)
    
df['Title'] = t

f = []
for i in df['FirstName']:
    try:
        c= re.sub("^\s+|\s+$", "", i, flags=re.UNICODE)
        c= " ".join(re.split("\s+", c, flags=re.UNICODE))
        f.append(c)
    except:
        f.append(np.nan)
    
df['FirstName'] = t


l = []
for i in df['LastName']:
    try:
        d= re.sub("^\s+|\s+$", "", i, flags=re.UNICODE)
        d= " ".join(re.split("\s+", d, flags=re.UNICODE))
        l.append(d)
    except:
        l.append(np.nan)
    
df['LastName'] = l

s = []
for i in df['City']:
    try:
        e= re.sub("^\s+|\s+$", "", i, flags=re.UNICODE)
        e= " ".join(re.split("\s+", e, flags=re.UNICODE))
        s.append(e)
    except:
        s.append(np.nan)
    
df['City'] = e

x = []
for i in df['Country']:
    try:
        
        y= re.sub("^\s+|\s+$", "", i, flags=re.UNICODE)
        y= " ".join(re.split("\s+", y, flags=re.UNICODE))
        x.append(y)
    except:
        x.append(np.nan)
    
df['Country'] = x
```

```python
p = []
for i in df['PostalCode']:
    try:
        a= re.sub(r'[a-zA-Z,?|_ ]+', '', str(i))
        a = int(a)
        p.append(a)
    except:
        p.append(np.nan)
        
df['PostalCode'] = p
```

```python
p = []

for i in df['Cellphone']:
    try:
        a = int(i)
        p.append(a)
    except:
        p.append(np.nan)
    
df['Cellphone'] = p

p = []

for i in df['Telephone']:
    try:
        a = int(i)
        p.append(a)
    except:
        p.append(np.nan)
    
df['Telephone'] = p
```

```python
df1=df[pd.notnull(df['Telephone'])]
df2=df[pd.notnull(df['Cellphone'])]
```

```python
p = []
l = []
for i in df1['Telephone']:
    a =  str(i)
    a = a.split('.')[0]
    b = a [-9:]
    c = len(b)
    p.append(b)
    l.append(c)
    
df1['Telephone'] = p
df1['CellLength']  = l

df1 = df1.loc[df1['CellLength']==9,:'Source']
df1['Telephone'] = df1['Telephone'].astype(int)

t = []
for i in df1['Telephone']:
    if i <400000000 or i>500000000:
        t.append('Telephone')
    else:
        t.append('Cellphone')

df1['Type'] = t

df4= df1.loc[df1['Type']=='Cellphone',:'Source']
df3= df1.loc[df1['Type']=='Telephone',:'Source']
df4['Cellphone']= df4['Telephone']
df4['Telephone'] = np.nan
df1 = pd.concat([df3,df4])
```

```python
p = []
l = []
for i in df2['Cellphone']:
    a =  str(i)
    a = a.split('.')[0]
    b = a [-9:]
    c = len(b)
    p.append(b)
    l.append(c)
    
df2['Cellphone'] = p
df2['CellLength']  = l

df2=df2.loc[df2.CellLength==9,:'Source']
df2['Cellphone']= df2['Cellphone'].astype(int)

t = []
for i in df2['Cellphone']:
    if i <400000000 or i>500000000:
        t.append('Telephone')
    else:
        t.append('Cellphone')

df2['Type'] = t


df5= df2.loc[df2['Type']=='Telephone',:'Source']
df6= df2.loc[df2['Type']=='Cellphone',:'Source']

df5['Telephone']= df5['Cellphone']
df5['Cellphone'] = np.nan
df2 = pd.concat([df5, df6])
```

```python
final = pd.concat([df1,df2])
```

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("NSW 01.08.csv", encoding='latin')
df2 = pd.read_csv("NSW 01.08 (1).csv", encoding='latin')
df3 = pd.read_csv("NSW 01.08 CSV.csv", encoding='latin')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
df = pd.concat([df1,df2,df3])
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop_duplicates('Phone',inplace=True)
```

```python
df.shape
```

```python
df = df.iloc[1:,:]
```

```python
p = []
l = []
for i in df['Phone']:
    a =  str(i)
    a = a.split('.')[0]
    b = a [-9:]
    c= len(b)
    p.append(b)
    l.append(c)
    
df['Phone'] = p
df['CellLength']  = l
```

```python
df.head()
```

```python
df['CellLength'].value_counts()
df = df.loc[df['CellLength']==9,:'Postcode']
```

```python
df.shape
```

```python
df.drop_duplicates('Phone', inplace=True)
```

```python
df.shape
```

```python
df['Postcode'] = df.Postcode.astype('int64')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.rename(columns = {'First/Full':'FirstName', 'Add1':'StreetAddress','Add2':'City', 'Add3':'Country', 
                     'Surname':'LastName', 'Phone':'Cellphone','Postcode':'PostalCode'}, inplace=True)
```

```python
df['Title'] = df.apply(lambda _: '', axis= 1)
df['Source'] = 'NSW 01.08'
```

```python
df['Cellphone'] = df.Cellphone.astype('int64')
```

```python
t = []
for i in df['Cellphone']:
    if i <400000000 or i>500000000:
        t.append('Telephone')
    else:
        t.append('Cellphone')

df['Type'] = t
```

```python
df.head()
```

```python
df1= df.loc[df['Type']=='Cellphone',:'Source']
df2= df.loc[df['Type']=='Telephone',:'Source']
```

```python
df2.rename(columns={'Cellphone':'Telephone'},inplace=True)
```

```python
df2['Cellphone'] = df2.apply(lambda _: '', axis= 1)
```

```python
df2 = df2[['Title', 'FirstName', 'LastName', 'StreetAddress', 'City', 'Country','PostalCode', 'Cellphone', 'Telephone', 'Source']]
```

```python
df1['Telephone'] = df1.apply(lambda _: '', axis= 1)
```

```python
df1 = df1[['Title', 'FirstName', 'LastName', 'StreetAddress', 'City', 'Country','PostalCode', 'Cellphone', 'Telephone', 'Source']]
```

```python
df = pd.concat([df1,df2])
```
